```python
import json
def read_json(path):
    with open(path, encoding="utf-8") as f: 
        return json.load(f)
def write_json(path, data):
    with open(path, 'w', encoding="utf-8") as f: 
        return json.dump(data, f, indent=2)
```

```python
data = read_json('fifa18.json')
```
