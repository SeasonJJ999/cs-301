```python
import pandas as pd
from pandas import Series, DataFrame
import sqlite3
```

```python
c = sqlite3.connect("bus.db")
```

```python
def qry(sql):
    return pd.read_sql(sql, c)
```

```python
qry("""
SELECT * 
FROM boarding
ORDER BY Route
""")
```

# How many people use the bus system daily?

```python
qry("""
SELECT SUM(DailyBoardings) FROM boarding
""")
```

# How many stops does each route have?

```python
qry("""
SELECT COUNT(), Route
FROM boarding
GROUP BY Route
""")
```

# Which routes have less than 20 stops?

```python
qry("""
SELECT COUNT() as stops, Route
FROM boarding
GROUP BY Route
HAVING stops < 20
""")
```

# How many people board each route daily?

Show most popular routes first.

```python
qry("""
SELECT SUM(DailyBoardings) as people, Route
FROM boarding
GROUP BY Route
""")
```

# Which bus routes have at least 3 stops with more than 50 daily boardings?

```python
qry("""
SELECT SUM(DailyBoardings) as people, Route
FROM boarding
GROUP BY Route
HAVING people >= 3
""")
```

# Which bus routes have at least 10 stops with more than 50 daily boardings?

```python
qry("""
SELECT COUNT() as stops, SUM(DailyBoardings) as people, Route
FROM boarding
GROUP BY Route
HAVING people > 50 AND stops >=10
""")
```

# Which bus routes have at least 3 stops with more than 100 daily boardings?

```python
qry("""
SELECT COUNT() as stops, SUM(DailyBoardings) as people, Route
FROM boarding
GROUP BY Route
HAVING people > 100 AND stops >=3
""")
```
